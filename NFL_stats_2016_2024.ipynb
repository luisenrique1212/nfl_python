Creation of dictionary to generate column to load the location of each team

```python
dict_teams = {'ARI':'Arizona, Phoenix','ATL':'Atlanta, Georgia','BAL':'Baltimore, Maryland','BUF':'Buffalo, New York',
'CAR':'Charlotte, North Carolina','CHI':'Chicago, Illinois','CIN':'Cincinnati, Ohio','CLE':'Cleveland, Ohio',
'DAL':'Arlington, Texas','DEN':'Denver, Colorado','DET':'Detroit, Michigan','GB':'Green Bay, Wisconsin','HOU':'Houston, Texas',
'IND':'Indianapolis, Indiana','JAX':'Jacksonville, Florida','KC':'Kansas City, Missouri','LA':'Inglewood, California',
'LAC':'Costa Mesa, California','LV':'Paradise, Nevada','MIA':'Miami, Florida','MIN':'Minneapolis, Minnesota','NE':'Foxborough, Massachusetts',
'NO':'New Orleans, Louisiana','NYG':'East Rutherford, New Jersey','NYJ':'East Rutherford, New Jersey','PHI':'Philadelphia, Pennsylvania',
'PIT':'Pittsburgh, Pennsylvania','SEA':'Seattle, Washington','SF':'Santa Clara, California','TB':'Tampa, Florida','TEN':'Nashville, Tennessee',
'WAS':'Washington, D.C.'}
```

Function to load the column of location to each team

```python
def home_location(pbp_temp):
  pbp_temp_yards_l = []
  for index in pbp_temp['home_team']:
    for index_d in dict_teams.keys():
      if index == index_d:
        pbp_temp_yards_l.append(dict_teams.get(index_d))
  return pbp_temp_yards_l
```

Function to load the year of each season

```python
def season(pbp_temp,season):
  pbp_temp_yards_s = []
  for index in pbp_temp['home_team']:
    pbp_temp_yards_s.append(season)
  return pbp_temp_yards_s
```

Use the function to add the location and season year to each record

```python
pbp_temp_yards_2016['home_location_team'] = home_location(pbp_temp_yards_2016)
pbp_temp_yards_2017['home_location_team'] = home_location(pbp_temp_yards_2017)
pbp_temp_yards_2018['home_location_team'] = home_location(pbp_temp_yards_2018)
pbp_temp_yards_2019['home_location_team'] = home_location(pbp_temp_yards_2019)
pbp_temp_yards_2020['home_location_team'] = home_location(pbp_temp_yards_2020)
pbp_temp_yards_2021['home_location_team'] = home_location(pbp_temp_yards_2021)
pbp_temp_yards_2022['home_location_team'] = home_location(pbp_temp_yards_2022)
pbp_temp_yards_2023['home_location_team'] = home_location(pbp_temp_yards_2023)
pbp_temp_yards_2024['home_location_team'] = home_location(pbp_temp_yards_2024)
pbp_temp_yards_2016['season'] = season(pbp_temp_yards_2016,2016)
pbp_temp_yards_2017['season'] = season(pbp_temp_yards_2017,2017)
pbp_temp_yards_2018['season'] = season(pbp_temp_yards_2018,2018)
pbp_temp_yards_2019['season'] = season(pbp_temp_yards_2019,2019)
pbp_temp_yards_2020['season'] = season(pbp_temp_yards_2020,2020)
pbp_temp_yards_2021['season'] = season(pbp_temp_yards_2021,2021)
pbp_temp_yards_2022['season'] = season(pbp_temp_yards_2022,2022)
pbp_temp_yards_2023['season'] = season(pbp_temp_yards_2023,2023)
pbp_temp_yards_2024['season'] = season(pbp_temp_yards_2024,2024)
```

```python
pbp_temp_yards = pd.concat([pbp_temp_yards_2016, pbp_temp_yards_2017, pbp_temp_yards_2018, pbp_temp_yards_2019, pbp_temp_yards_2020, pbp_temp_yards_2021, pbp_temp_yards_2022, pbp_temp_yards_2023, pbp_temp_yards_2024])
```
